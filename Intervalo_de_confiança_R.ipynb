## Atividade | Estatistica Computacional em R


Consumo Médio de carros (mtcars$mpg)

Contexto. Uma revista afirma que carros dessa amostra fazem, em média, 20 mpg.

Objetivos.

1. Verificar se os dados seguem uma distribuição normal (teste Shapiro)
2. Construir IC 95% para a média de mpg (construa IC para 90% e 99% e descreva seus sentimentos).
3. Testar H₀: μ = 20 (bilateral).

```r
data(mtcars)
```

```r
head(mtcars)
```

```r
mpg_total <- mtcars$mpg
```

1. Verificar se os dados seguem uma distribuição normal:


*   Hipótese Nula (H0):	A amostra foi extraída de uma população com distribuição normal.
*   Hipótese Alternativa (H1):	A amostra não foi extraída de uma população com distribuição normal.

Se p>0,05 (maior que o nível de significância): Não rejeitamos H0. Com um nível de confiança de 95%

Concluímos que não há evidência estatística suficiente para dizer que os dados não seguem uma distribuição normal (ou seja, assumimos a normalidade).




```r
shapiro.test(mpg_total)  #Teste de Shapiro
```

```r
qqnorm(mpg_total) # Cria o Q-Q Plot
qqline(mpg_total) # Adiciona a linha teórica de normalidade
```

Considerando p-value = 0.1229 > 0,05 podemos assumir a normalidade da amostra.

2. Construir IC 95% para a média de mpg (construa IC para 90% e 99% e descreva seus sentimentos).

```r
media_mpg <- mean(mpg_total)
print(media_mpg)
```

```r
ic_95 = t.test(mpg_total, conf.level = 0.95)$conf.int
ic_90 = t.test(mpg_total, conf.level = 0.90)$conf.int
ic_99 = t.test(mpg_total, conf.level = 0.99)$conf.int

cat("Média:", media_mpg, "\n")
cat("IC 95%:", ic_95, "\n")
cat("IC 90%:", ic_90, "\n")
cat("IC 99%:", ic_99, "\n")
```

Ao calcular os diferentes níveis de confiança é possível notar que quanto maior o nível de confiança maior também é o intervalo, ou seja, maior a incerteza sobre onde está a média verdadeira. Aumentando o intervalo de 95% para 99% é possível notar que o intervalo se amplia para ter certeza que a média real está sendo contemplada nesse intervalo.


3. Testar H₀: μ = 20 (bilateral).

```r
# teste t bilateral contra H0 = 20
t.test(mpg_total, mu = 20, alternative = "two.sided")
```

Estatística do teste: t=0.08506

Graus de liberdade: df=31

p-value = 0.9328

IC = 95%

Sendo o p-value > 0.05 não rejeitamos a hipóteste nula, não tendo evidência suficiente para dizer que a média é diferente de 20.

*Dica: se a média está dentro do intervalo de confiança, normalmente não se rejeita a hipótese nula*
